### Data Cleaning

Using the pandas library in python, we are importing csv files as DataFrames and combining them into a larger dataset to use for visualizations and analysis.

```python
import pandas as pd
```

#### Tax Rates 2005-2022

```python
df_tax = pd.read_csv("data05-22/taxrates.csv")
df_tax = df_tax.round(2) 
df_tax = df_tax.drop(columns=['2023']) #dropping 2023 to keep in line with remaining datasets and for easier merge later on.
df_tax
```

We see that District of Columbia is at the bottom. To maintain consistency, we are sorting all the datasets alphabetically by state so District of Columbia should come after Delaware

```python
df_tax = df_tax.sort_values(by=['State']) #
df_tax = df_tax.reset_index(drop=True)
df_tax.head()
```

#### Business Applications 2005-2022

```python
df_biz = pd.read_csv('data05-22/biz_apps_new.csv')
df_biz.head()
```

#### Population by State 2005-2022

```python
df_pop = pd.read_csv("data05-22/population_bystate.csv")
df_pop.head()
```

#### GDP of States 2005-2022

```python
df_gdp = pd.read_csv("data05-22/GDP_bystate.csv")
df_gdp.head()
```

#### Minimum Wage 2005-2022

```python
df_mw = pd.read_csv("data05-22/min_wage.csv")
df_mw = df_mw.drop(columns=['2023']) #dropping 2023 for maintaining consistency
df_mw.head()
```

#### Unemployment Rate 2005-2022

```python
df_ur = pd.read_csv('data05-22/Unemployment_rates.csv')
df_ur.head()
```

#### Median Household Incomes 2005-2022

```python
df_mhi = pd.read_csv('data05-22/median_household_income.csv')
df_mhi = df_mhi.sort_values(by=['State']).reset_index(drop=True)
df_mhi.head()
```

### Merging into single dataset

```python
#adding tax
df = df_tax.melt(id_vars= ['State'])
df = df.rename(columns={"variable":"year", 'value':'tax_rate'})
df = df.sort_values(by=['State','year'])
df = df.reset_index(drop=True)
df.head()
```

```python
#adding business applications
df_ba = df_biz.melt(id_vars = ['Name', 'State'])
df_ba = df_ba.rename(columns={"variable":"year", 'value':'business_applications'})
df_ba = df_ba.sort_values(by=['State','year']).reset_index(drop=True)

df['business_applications'] = df_ba['business_applications']
```

```python
#adding GDP
df_gd = df_gdp.melt(id_vars = ['State'])
df_gd = df_gd.rename(columns={"variable":"year", 'value':'gdp'})
df_gd = df_gd.sort_values(by=['State','year']).reset_index(drop=True)

df['gdp'] = df_gd['gdp']
```

```python
#adding minimum wage
df_mwage = df_mw.melt(id_vars=['State'])
df_mwage = df_mwage.rename(columns={"variable":"year", 'value':'min_wage'})
df_mwage = df_mwage.sort_values(by=['State','year']).reset_index(drop=True)
df['min_wage'] = df_mwage['min_wage']
```

```python
#adding unemployment
df_ur_new = df_ur.melt(id_vars=['State'])
df_ur_new = df_ur_new.rename(columns={"variable":"year", 'value':'unemployment_rate'})
df_ur_new = df_ur_new.sort_values(by=['State','year']).reset_index(drop=True)

df['unemployment_rate'] = df_ur_new['unemployment_rate']
```

```python
#adding population
df_pp = df_pop.melt(id_vars=['State'])
df_pp = df_pp.rename(columns={"variable":"year", 'value':'population'})
df_pp = df_pp.sort_values(by=['State','year']).reset_index(drop=True)

df['population'] = df_pp['population']
```

```python
#adding median household income
df_m = df_mhi.melt(id_vars=['State'])
df_m = df_m.rename(columns={"variable":"year", 'value':'median_income'})
df_m = df_m.sort_values(by=['State','year']).reset_index(drop=True)

df['median_hh_income'] = df_m['median_income']
```
