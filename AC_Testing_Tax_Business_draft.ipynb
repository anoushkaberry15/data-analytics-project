# Data Preparation 
## Tax Rates

```python
# Import Tax rate data 
import pandas as pd
tax_rates = pd.read_csv(r'/Users/alex/Documents/Data Science Master/First Semester/Group Project Analytics in Business/Tax rates by state 2000-2023.csv', index_col = 0)
```

```python
# Remove Na's from dataset
tax_rates = tax_rates.dropna()
```

```python
# Round tax rates to second decimal
tax_rates = tax_rates.round(2)
tax_rates
```

```python
# Drop columns excluded from analyses (2001 to 2005 and 2022, 2023)
# These years were selected to take different datasets into account
tax_rates = tax_rates.drop(columns=['2000', '2001', '2002', '2003', '2004', '2022', '2023'])

tax_rates
```

```python
# Calculate change in tax rates over time and save in new dataframe
tax_rate_change = tax_rates.iloc[:, 0:].diff(axis = 1)

# Drop 2005 column as no change in tax rate in first year of data can be calculated
tax_rate_change = tax_rate_change.drop(columns = ['2005'])
tax_rate_change

# Save tax rate change for 5 states with largest change
# Introduce new column in dataframe calculating absolute change over time
tax_rate_change['Abs_Change'] = tax_rate_change.iloc[:, 0:].abs().max(axis = 1)

# Save 5 states with highest change in tax rate

tr_change_largest = tax_rate_change.sort_values(by='Abs_Change', ascending=False).head(5)
tr_change_largest 
```
